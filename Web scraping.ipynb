```python
from autoscraper import AutoScraper
```

```python
url='https://clutch.co/web-developers/india'
```

```python
wanted_list = ["Dreamztech Solutions","4.9","Kolkata, India", "< $25 / hr", "$10,000+","50 - 249","https://dreamztechusa.com/?utm_source=clutch.co&utm_medium=referral&utm_campaign=web-developers-india"]
```

```python
scraper = AutoScraper()
result = scraper.build(url, wanted_list)
print(result)
```

```python
scraper.get_result_similar("https://clutch.co/web-developers/india",grouped=True)
```

```python
scraper.set_rule_aliases({'rule_95wo': 'Company', 'rule_jwbn': 'rating' , 'rule_xrhs' :'Location' , 'rule_yjrz' : 'Hourly Rate','rule_ivez' :'Min Project Size','rule_psj0' :'Employee Size','rule_aw0s':'Website'})
scraper.keep_rules(['rule_95wo', 'rule_jwbn','rule_xrhs','rule_yjrz','rule_ivez' ,'rule_psj0' ,'rule_aw0s'])
scraper.save('web_scrape')
```

```python
scraper.load('web_scrape')
```

```python
result=scraper.get_result_similar("https://clutch.co/web-developers/india",group_by_alias=True)
```
